```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = pd.rolling_mean(timeseries, window=24)
    rolstd = pd.rolling_std(timeseries, window=24)

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean Forecast')
    std = plt.plot(rolstd, color='black', label = 'Stdev')
    plt.legend(loc='best')
    plt.title('ARMA & Standard Deviation')
    plt.show(block=False)
    

test_stationarity(f1_9['Load_1'])


```

```python
#ARIMA 
from pandas import read_csv
from pandas import datetime
from pandas import DataFrame
from statsmodels.tsa.arima_model import ARIMA
from matplotlib import pyplot
import numpy as np
from sklearn.metrics import mean_squared_error
from math import sqrt


# fit model via Grid Search
def search_Arima(data, N_ar,N_i,N_ma):
    mses = []
    model_s = []
    actual = np.exp(np.float64(data))
    for i in range(1,N_ar):
        for j in range(1,N_ma):
            model = ARIMA(np.float64(data), order=(i,N_i,j))
            model_fit = model.fit(disp=0)
            y_hat = model_fit.predict() 
            y_hats = np.exp(y_hat)
            mse = sqrt(mean_squared_error(actual, np.float64(y_hats)))
            mses.append(mse)
            model_s.append('AR = {}, I = {}, MA = {}'.format(i,N_i,j))
    return model_s,mses
            

search_Arima(nts1, 5,0,9)
```
